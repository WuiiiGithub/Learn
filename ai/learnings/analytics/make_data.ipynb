```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('../../data/heart_disease.csv').select_dtypes(include=['number'])
df.head()
```

```python
renamer = {}
for col in df.columns:
    name = input(f"Old: {col} || New: ")
    renamer[col] = name
    renamer[name] = col
```

```python
df.rename(columns=renamer, inplace=True)
df
```

```python
df.info()
```

```python
imputer = KNNImputer(n_neighbors=5)
cols = df.columns
df = imputer.fit_transform(df)
df = pd.DataFrame(df, columns=cols)
df
```

```python
df.info()
```

```python
df.to_csv('../../data/cholestrol.csv')
```
