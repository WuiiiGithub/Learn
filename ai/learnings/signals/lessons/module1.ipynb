## Visualising Basic Waveforms

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(-3, 3, 10000)
sine_wave = np.sin(2 * np.pi * 1 * time)

plt.scatter(time, sine_wave, marker='.')
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("sin(2$\\pi$ft)")
plt.title("A sine wave")
plt.show()
```

```python
time = np.linspace(-3, 3, 10000)
cos_wave = np.cos(2 * np.pi * 1 * time)

plt.scatter(time, cos_wave, marker='.')
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("cos(2$\\pi$ft)")
plt.title("A cos wave")
plt.show()
```

```python
time = np.linspace(-3, 3, 10000)
sine_wave = 0.25*np.sin(2 * np.pi * 1 * time)


plt.scatter(time, sine_wave, marker='.')
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("sin(2$\\pi$ft)")
plt.title("A sine wave")
plt.show()
```

```python
time = np.linspace(-3, 3, 10000)
sine_wave = 0.25*np.sin(2 * np.pi * 1 * time) + 0.5


plt.scatter(time, sine_wave, marker='.')
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("sin(2$\\pi$ft)")
plt.title("A sine wave")
plt.show()
```

```python
time = np.linspace(-3, 3, 10000)
sine_wave = np.sin(2 * np.pi * 1 * time)
square_wave = np.sign(sine_wave)

plt.scatter(time, square_wave, marker='.')
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("sign(sin(2$\\pi$ft))")
plt.title("A square wave")
plt.show()
```

```python
time = np.linspace(-3, 3, 10000)
sine_wave = np.sin(2 * np.pi * 1 * time)
square_wave = np.sign(sine_wave)

plt.plot(time, square_wave)
plt.yticks([-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1])
plt.xlabel("time")
plt.ylabel("sign(sin(2$\\pi$ft))")
plt.title("A square wave")
plt.show()
```

```python
shifted = np.roll(sine_wave, 10000//12)
scaled = 2 * sine_wave

plt.figure(figsize=(12, 4))
plt.plot(time, sine_wave, label='Original')
plt.plot(time, shifted, label='Time-Shifted')
plt.plot(time, scaled, label='Scaled (2x)')
plt.legend()
plt.title('Time & Amplitude Transformations')
plt.grid(True)
plt.show()
```

```python
transformed = np.flip(sine_wave)

plt.figure(figsize=(12, 4))
plt.plot(time, sine_wave, label='Original')
plt.plot(time, transformed, label='Time-Reversed')
plt.legend()
plt.title('Time Reversal Transformations')
plt.grid(True)
plt.show()
```

```python
transformed = np.clip(sine_wave, -0.5, 0.5)

plt.figure(figsize=(12, 4))
plt.plot(time, sine_wave, label='Original')
plt.plot(time, transformed, label='Clipped Transform')
plt.legend()
plt.title('Clipped Transformations')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(time, sine_wave, label='Original', color='lightgreen')
plt.plot(time[::50], sine_wave[::50], '--', color='orange', linewidth=2, label='Signal with 50')
plt.plot(time[::50], sine_wave[::50], 'o', color='red', markersize=5)
plt.title('Downsample x50')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(time, sine_wave, label='Original', color='lightgreen')
plt.plot(time[::100], sine_wave[::100], '--', color='orange', linewidth=2, label='Signal with 100')
plt.plot(time[::100], sine_wave[::100], 'o', color='red', markersize=5)
plt.title('Downsample x100')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(time, sine_wave, label='Original', color='lightgreen')
plt.plot(time[::500], sine_wave[::500], '--', color='orange', linewidth=2, label='Signal with 500')
plt.plot(time[::500], sine_wave[::500], 'o', color='red', markersize=5)
plt.title('Downsample x500')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(time, sine_wave, label='Original', color='lightgreen')
plt.plot(time[::1000], sine_wave[::1000], '--', color='orange', linewidth=2, label='Signal with 1000')
plt.plot(time[::1000], sine_wave[::1000], 'o', color='red', markersize=5)
plt.title('Downsample x1000')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
